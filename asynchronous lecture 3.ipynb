```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
y = np.linspace(0,1,num=21)
#Values from Clayton 1983, table 2-4, H gas, T~10^4K
Γ1 = np.array([1.6666,1.2097,1.1688,1.1537,1.1460,
               1.1415,1.1386,1.1368,1.1356,1.1349,
               1.1347,1.1349,1.1356,1.1368,1.1386,
               1.1415,1.1460,1.1537,1.1688,1.2097,1.6666])
```

```python
fig, ax = plt.subplots()
ax.plot(y, Γ1)
i = np.argmin(Γ1)
ax.axvline(y[i], linestyle='dashed')
ax.set_xlabel('ionization fraction')
ax.set_ylabel('$\Gamma_1$')
ax.set_ylim(1, 1.6666)
ax.plot(y[i], Γ1[i], marker='o')
ax.text(y[i]+0.05, 1.05, "y={}, $\Gamma_1$={}".format(y[i], Γ1[i]))
```

After watching these lectures, solve the following:

In lecture, we learn that the system
\begin{equation}
\vec{\nabla}\cdot(K \vec{\nabla}T) = 0
\end{equation}
for non-constant $K$ and $\nabla T$ can be solved if $K=K(T)$.  Here we do so for an simplified version of the solar corona.

In the solar corona, thermal conduction is dominated by electrons.  The heat diffusion coefficient $K(T)$ scales as:
\begin{equation}
    K(T) \propto T^{5/2}
\end{equation}
this corresponds to heat conduction by un-magnetized electrons in a plasma, following Braginskii (1965) and the NRL plasma formulary.  A simplified version of this is sometimes called "Spitzer conductivities" in astrophysics, but Braginskii is the original source and includes magnetization and other important effects.

**Solve:**

1. Assuming a spherical geometry, and that T varies only in radius $r$, $T=T(r)$ and $K=K(T)=T^{5/2}$, write the expression $\vec{\nabla}\cdot(K \vec{\nabla}T) = 0$ in terms of T and r.

2. Integrate the expression you get in (1) twice to get T(r).  

3. You'll have two integration constants, and you'll need two boundary conditions to evaluate them:
     - assume that $T(r)\rightarrow0$ as $r\rightarrow\infty$
     - assume that $T(r) = T_0$ (a constant) at $r=r_0$ for some chosen reference radius $r_0$.

You should obtain an answer of the form:
\begin{equation}
    \left(\frac{T}{T_0}\right) = \left(\frac{r}{r_0}\right)^\alpha
\end{equation}
for some power-law index $\alpha$.  This is the solution to thermal equilibrium for a system dominated by electron conduction processes (in a spherical geometry).

4. Think but don't answer: are there any problems with this solution?
